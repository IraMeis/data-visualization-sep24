```python
from plotnine import *
from plotnine.data import *

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/Laptop-Price.csv')
```

```python
#    
ggplot(data=df) +\
geom_bar(mapping=aes(x='Company')) + theme(figure_size = (12, 5))
```

```python
#   / 
ggplot(data=df) +\
geom_bar(mapping=aes(x='Company', fill='OpSys'), position="dodge") + theme(figure_size = (12, 5))
```

```python
#  RAM /  
ggplot(data=df) +\
geom_bar(mapping=aes(x='TypeName', fill='Ram'), position="dodge") + theme(figure_size = (9, 5))
```

```python
#   / 
ggplot(data=df) +\
geom_point(mapping=aes(x='Price_euros', y='Inches'))
```

```python
# boxplot   
ggplot(data=df) +\
geom_boxplot(mapping=aes(x='Company', y='Price_euros')) + theme(figure_size = (12, 5))
```

```python
#      
ggplot(data=df) +\
geom_smooth(mapping=aes(x='Inches', y='Price_euros', group=1))
```

```python
#     SSD
ggplot(data=df) +\
geom_smooth(mapping=aes(x='SSD', y='Price_euros', group=1))
```

```python
#   ( / )
ggplot(data=df) +\
geom_point(mapping=aes(x='Inches', y='ScreenResolution'))
```

```python
# GPU providers
ggplot(data=df) +\
geom_bar(mapping=aes(x='Gpu Brand'))
```

```python
# GPU providers  CPU providers
ggplot(data=df) +\
geom_bar(mapping=aes(x='Cpu Brand', fill='Gpu Brand'), position="dodge")
```
